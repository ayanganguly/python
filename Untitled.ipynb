```python
import pandas as pd
import numpy as np
import joblib

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, mean_squared_error, r2_score

from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression, ElasticNet
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("BALANCED_learning_dataset_10k.csv")
df.head()
```

```python
print(df.shape)
print(df.describe())

plt.hist(df['performance_score'], bins=40)
plt.title("Performance Score Distribution")
plt.show()
```

```python
FEATURES = [
    'prior_knowledge_score','avg_quiz_score','cognitive_load_score','avg_time_per_question',
    'error_rate','revision_frequency','help_requests_count','attention_span_score',
    'mental_fatigue_score','time_variation','concept_mastery_rate','engagement_rate',
    'avg_study_time_per_day'
]

X = df[FEATURES]

y_path = df['learning_path_type']
y_ready = df['ready_for_advanced']
y_ready_svc = df['ready_for_advanced_svc']
y_diff = df['recommended_difficulty']
y_perf = df['performance_score']
```

```python
X_train, X_test, \
y_path_train, y_path_test, \
y_ready_train, y_ready_test, \
y_ready_svc_train, y_ready_svc_test, \
y_diff_train, y_diff_test, \
y_perf_train, y_perf_test = train_test_split(
    X,
    y_path,
    y_ready,
    y_ready_svc,
    y_diff,
    y_perf,
    test_size=0.2,
    random_state=42
)
```

```python
clf_tree = DecisionTreeClassifier(max_depth=6, min_samples_leaf=10, random_state=42)
clf_tree.fit(X_train, y_path_train)

print("Decision Tree Performance")
print(classification_report(y_path_test, clf_tree.predict(X_test)))

joblib.dump(clf_tree, "decision_tree_learning_path.pkl")
```

```python
pipe_log = Pipeline([
    ('scaler', StandardScaler()),
    ('logreg', LogisticRegression(max_iter=2000))
])

pipe_log.fit(X_train, y_ready_train)

print("Logistic Regression Performance")
print(classification_report(y_ready_test, pipe_log.predict(X_test)))

joblib.dump(pipe_log, "logreg_ready.pkl")
```

```python
clf_rf = RandomForestClassifier(
    n_estimators=150,
    max_depth=12,
    min_samples_leaf=5,
    random_state=42
)

clf_rf.fit(X_train, y_diff_train)

print("Random Forest Performance")
print(classification_report(y_diff_test, clf_rf.predict(X_test)))

joblib.dump(clf_rf, "rf_difficulty.pkl")
```

```python
pipe_svc = Pipeline([
    ('scaler', StandardScaler()),
    ('svc', SVC(probability=True, kernel='rbf'))
])

pipe_svc.fit(X_train, y_ready_svc_train)

print("SVC Performance")
print(classification_report(y_ready_svc_test, pipe_svc.predict(X_test)))

joblib.dump(pipe_svc, "svc_ready.pkl")
```

```python
pipe_en = Pipeline([
    ('scaler', StandardScaler()),
    ('en', ElasticNet(alpha=0.1, l1_ratio=0.5, max_iter=5000))
])

pipe_en.fit(X_train, y_perf_train)

pred_perf = pipe_en.predict(X_test)

print("ElasticNet MSE:", mean_squared_error(y_perf_test, pred_perf))
print("ElasticNet R2:", r2_score(y_perf_test, pred_perf))

joblib.dump(pipe_en, "elasticnet_performance.pkl")
```

```python
sample = X_test.iloc[0:1]

print("Decision Tree Path:", clf_tree.predict(sample)[0])
print("Difficulty:", clf_rf.predict(sample)[0])
print("Ready (LogReg):", pipe_log.predict(sample)[0])
print("Ready (SVC):", pipe_svc.predict(sample)[0])
print("Performance Score:", pipe_en.predict(sample)[0])
```
